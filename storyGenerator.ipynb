```python
import random

def setting_the_scene():
    """Ask a player to choose a story"""
    setting = {
        1: "forest",
        2: "castle",
        3: "space"
    }
    print("Choose your preferred setting")
    for key, value in setting.items():
         print(f"{key}. {value.capitalize()}")
            
    while True:
        try:
            select_setting = int(input("Enter a number that correlates to your setting choice: "))
            if select_setting in setting:
                print(f"You are in a {setting[select_setting]}.")
                return setting[select_setting]
            else:
                print("Invalid setting choice, Please choose a number from 1 to 3")
        except ValueError:
            print("Invalid input. Please enter a number.")

    
def character_creation():
    """Asking player to choose a character"""
    character = {
        1:"hero",
        2:"villain"
    }
    print("Choose a character")
    for key, value in character.items():
        print(f"{key}. {value.capitalize()}")
        
    while True:
        try:
            select_character = int(input("Enter a number that correlates to your character choice: "))
            if select_character in character:
                print(f"You are a {character[select_character]}.")
                return character[select_character]
            else:
                print("Invalid selection, Please choose a number from 1 to 2")
        except ValueError:
            print("Invalid input. Please enter a number.")


def plot_development():
    """Develop a basic plot with different events or challenges"""
    events = [
        "A cellphone is found inside a sealed Egyptian tomb. What do you do?\n1. Keep it\n2. Ignore It ",
        "A plane takes off with 81 passengers. It lands with 82. What do you do?\n1. Question it\n2. Report to the airline",
        "As a joke, you use your cellphone to dial your own number. Someone picks up. How will you react?\n1. Scream\n2. Change your sim",
        "While on a leisurely stroll through the forest, you find strange footprints. Do you follow them?\n1. I will follow them\n2. I will redirect from the path",
        "You briefly go out in the snow and come back inside. When you go out again, your footprint from earlier leads in a different direction. What do you do?\n1. Weary\n2. Scream"
    ]
    decisions = []
    for event in events:
        print(event)
        while True:
            try:
                decision = input("Enter 1 or 2 to make your decision: ")
                if decision not in ['1', '2']:
                    raise ValueError("Invalid decision. Choose 1 or 2")
                else:
                    decisions.append(decision)
                    break
            except ValueError as e:
                print(e)
    return decisions
        
def possible_outcome():
    """Variables to represent possible outcomes"""
    outcomes = {
        "happy ending":["You have successfully completed your trivia quest and achieved a happy ending. Good job!", "Congratulations!"],
        "tragic ending":["Unfortunately, your journey took a tragic turn. Better luck next time!"]
    }
    closing_outcomes = random.choice(list(outcomes.keys()))
    print(outcomes[closing_outcomes][0])
    if len(outcomes[closing_outcomes]) > 1:
        print(outcomes[closing_outcomes][1])
    else:
        print("No additional message for this outcome.")
    return outcomes[closing_outcomes][0]

def write_to_file(story_result):
    """Writing story result to a file"""  
    try:
        with open('story.txt','w') as filename:
            filename.write(story_result)
        print("Your story has been saved to a text file")
    except Exception as e:
        print("An error occurred while writing to file:", e)
    
def read_content():
    """Reading story or exiting """
    try:
        read_story = input("Enter Yes to read the story or No to exit: ")
        if read_story.lower() == 'yes':
            with open('story.txt','r') as f:
                print(f.read())
        elif read_story.lower() == 'no':
            exit()
        else:
            print("Invalid input. Please enter Yes or No.")
    except Exception as e:
        print("An error occurred while reading file:", e)
                    
                                      
def thank_player():
    """Thank the player for playing and encourage them"""
    print("\nThank you for partaking. Please try different choices in subsequent playthroughs.")
    
    
def main():
    """Main function to run the interactive story generator."""
    try:
        print("Welcome to the Interactive Story Generator!")
        player_name = input("Please enter your name: ")

        scene = setting_the_scene()
        role = character_creation()

        print(f"Welcome, {player_name}! Let's begin your adventure as the {role} in the {scene}.")

        print("\nPlot Development:")
        story_result = plot_development()

        print("\nOutcome:")
        outcome = possible_outcome()

        write_to_file(outcome)

        read_content()

        thank_player()
    except Exception as e:
        print("An error occurred:", e)
    
if __name__ == "__main__":
    main()
```
