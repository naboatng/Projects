```python
print("******* STUDENT GRADE CALCULATOR ********")
print("Calculating student grades")
print()

def input_subjects():
    try:
        num_subjects = int(input("Enter the number of subjects: "))
        subjects_grades = {}
        student_name = input("Enter your full name: ")
        student_id = input("Enter your 4-digit ID number: ")

        # Validate the ID number length
        while len(student_id) != 4 or not student_id.isdigit():
            print("Invalid ID number. Please enter a 4-digit numeric ID.")
            student_id = input("Enter your 4-digit ID number:")

        for i in range(num_subjects):
            subject = input("Enter subject name: ")
            if subject.isalpha():   
                grade = float(input(f"Enter grade for {subject}: "))
                subjects_grades[subject] = grade
         
        return subjects_grades, student_name, student_id
    except ValueError:
        print("Invalid input. Please enter a valid number.")

# Overall Grade Calculation Function with Error Handling
def calculate_overall_grade(subjects_grades):
    try:
        total_grade = sum(subjects_grades.values())
        overall_average = total_grade / len(subjects_grades)

        # Letter Grade Assignment
        if 90 <= overall_average <= 100:
            letter_grade = 'A'
        elif 80 <= overall_average <= 90:
            letter_grade = 'B'
        elif 70 <= overall_average <= 80:
               letter_grade = 'C'
        elif 60 <= overall_average <= 70:
                letter_grade = 'D'
        else:
            letter_grade = 'F'

        return overall_average, letter_grade
    
    except ZeroDivisionError:
        print("Error: Division by zero. Please enter at least one subject.")

def main():
    try:
        subjects_grades, student_name, student_id = input_subjects()
        overall_average, letter_grade = calculate_overall_grade(subjects_grades)

        # Display Results
        print("\nIndividual Grades: ")
        for subject, grade in subjects_grades.items():
            print(f"{subject}: {grade}")

        print(f"Student Name: {student_name}")
        print(f"Student ID: {student_id}")
        print(f"Overall Average Grade: {overall_average}")
        print(f"Letter Grade: {letter_grade}")

    except Exception as e:
        print(f"An error occurred: {e}")

# Run the program
if __name__ == "__main__":
    main()
```
